```python
#create GeoDataFrame for Chicago school districts

chicago = gp.GeoDataFrame.from_file('Chicago Public Schools - Geographic Networks/geo_export_35f1c1fa-243c-4dce-b1fd-e50e40557a60.shp')
```

```python
#set up geometry field for delay_geo.. right now, there are only lat/longs

s = gp.GeoSeries([Point(x,y) for x, y in zip(delays_geo['Longitude'], delays_geo['Latitude'])])
delays_geo['geometry'] = s
delays_geo.crs = {'init': 'epsg:4326', 'no_defs': True}
```

```python
#this method is used to create a heatmap given geometry points

def heatmap(d, bins=(100,100), smoothing=1.3, cmap='jet'):
    def getx(pt):
        return pt.coords[0][0]

    def gety(pt):
        return pt.coords[0][1]

    x = list(d.geometry.apply(getx))
    y = list(d.geometry.apply(gety))
    heatmap, xedges, yedges = np.histogram2d(y, x, bins=bins)
    extent = [yedges[0], yedges[-1], xedges[-1], xedges[0]]

    logheatmap = np.log(heatmap)
    logheatmap[np.isneginf(logheatmap)] = 0
    logheatmap = ndimage.filters.gaussian_filter(logheatmap, smoothing, mode='nearest')
    
    plt.imshow(logheatmap, cmap=cmap, extent=extent)
    plt.colorbar()
    plt.gca().invert_yaxis()
    plt.show()
```

```python
#create a heatmap of the Chicago Metra delays

heatmap(delays_geo['geometry'], bins = 50, smoothing = 1.3)
```

```python
#this method is given a school district and returns the total number of delays and the aggregate time of delays in seconds

def sum_delays(school_district):
    total = 0
    i = 0
    sum_time = 0
    while(i < len(delays_geo)):
        if(delays_geo['geometry'][i].within(school_district)==True):
            total = total + 1
            sum_time = sum_time + delays_geo['Delay'][i]
        i = i+1
    return(total, sum_time)
```

```python
#run the method sum_delays to get the number of delays and the aggregate time of delays for each school district

far_south_side_delays, far_south_side_delays_time = sum_delays(far_south_side)
midway_chicago_delays, midway_chicago_delays_time = sum_delays(midway_chicago)
englewood_delays, englewood_delays_time = sum_delays(englewood)
bronzeville_delays, bronzeville_delays_time = sum_delays(bronzeville)
pilsen_delays, pilsen_delays_time = sum_delays(pilsen)
mckinley_delays, mckinley_delays_time = sum_delays(mckinley)
chinatown_delays, chinatown_delays_time = sum_delays(chinatown)
humboldt_delays, humboldt_delays_time = sum_delays(humboldt)
saug_delays, saug_delays_time = sum_delays(saug)
belmont_delays, belmont_delays_time = sum_delays(belmont)
logan_delays, logan_delays_time = sum_delays(logan)
ravenswood_delays, ravenswood_delays_time = sum_delays(ravenswood)
chatham_delays, chatham_delays_time = sum_delays(chatham)
```
