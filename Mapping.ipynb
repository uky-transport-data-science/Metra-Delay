```python
import pandas as pd
routes = pd.read_csv('output1.csv')

# Extract the data
lat = routes['latitude'].values
lon = routes['longitude'].values
delay = routes['delay'].values
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# conda install basemap
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# 1. Draw the map background
fig = plt.figure(figsize=(15, 15))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=41.8781, lon_0=-87.6298,
            width=8E4, height=8E4)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter metra delay location data, with color reflecting length of delay
# the size can be variable, byt a standard size is set here
m.scatter(lon, lat, latlon=True,
          c=np.log10(delay), s=5,
          cmap='Reds', alpha=0.5)

# 3. create colorbar and legend
plt.colorbar(label=r'$\log_{10}({\rm delay})$')
plt.clim(1, 5)
```

```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create basemap layer
a = folium.Map([41.8781, -87.6298], zoom_start=11)
a
```

```python
#add markers for the location of data point collection
for index, row in routes.iterrows():
      if row["id"] != "South Pole Station":
        marker = folium.CircleMarker([row["latitude"], row["longitude"]], 
                                     radius=2,
                                     fill_color="#3db7e4")
        marker.add_to(a)
a.save('circ_map.html')
```

```python
# the heat map plugin accepts tuples so it has to be presented in that form
# TypeError: only length-1 arrays can be converted to Python scalars (doesn't like how lat long is being presented)
routeDelay = routes[['latitude', 'longitude']].as_matrix()

# plot heatmap
a.add_children(plugins.HeatMap(routeDelay, radius=5))
a
```
